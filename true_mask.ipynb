```python
import tifffile
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import glob
```

```python
c=0
files = glob.glob("tif_data/River/*.tif")
for file in files:
    n = (file.split('_')[2].split('.')[0])
    c+=1
    im = tifffile.imread(file)
    nir = im[:,:,5]/10
    r = im[:,:, 2]/10 
    ndwi=(nir-r)/(r+nir)
    #true_img = im[:,:, [3,2,1]]/10
    mask=np.where(ndwi>=0.2,1,0)
    im = Image.fromarray(np.uint8(mask*255),'L')
    im.save('tif_data/tif_mask/true_mask_{}.png'.format(n))
    if c%250==0:
            print(c)
```
